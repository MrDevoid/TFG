```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection  import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import sklearn.feature_selection
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
```

```python
audiosMezclados = pd.read_csv('sonidosMezclados.csv', na_values='--undefined--')
audiosAleman = pd.read_csv('sonidosAleman.csv', na_values='--undefined--')
audiosPelicula = pd.read_csv('sonidosPelicula.csv', na_values='--undefined--')
```

```python
X_Mezclados = audiosMezclados.drop(['Result'], axis=1)
y_Mezclados = audiosMezclados.Result

X_Aleman = audiosAleman.drop(['Result'], axis=1)
y_Aleman = audiosAleman.Result

X_Pelicula = audiosPelicula.drop(['Result'], axis=1)
y_Pelicula = audiosPelicula.Result
```

```python
X_MezcladosMean = X_Mezclados.mean(numeric_only=1) 
X_AlemanMean = X_Aleman.mean(numeric_only=1)
X_PeliculaMean = X_Pelicula.mean(numeric_only=1)
```

```python
for i in X_Mezclados.columns.values:
    X_Mezclados[i].replace(np.nan, X_MezcladosMean[i], inplace = True)
    X_Aleman[i].replace(np.nan, X_AlemanMean[i], inplace = True)
    X_Pelicula[i].replace(np.nan, X_PeliculaMean[i], inplace = True)
```

```python
plt.subplots(figsize=(10,10))
correlation = X_Mezclados.corr()
sns.heatmap(correlation, xticklabels = correlation.columns.values,yticklabels=correlation.columns.values)
plt.title('Feature correlations')
plt.show()
```

```python
scaler = MinMaxScaler()
scaled_Mezclados = scaler.fit_transform(X_Mezclados)
scaled_Aleman = scaler.fit_transform(X_Aleman)
scaled_Pelicula = scaler.fit_transform(X_Pelicula)
```

```python
X_MezcladosNor = pd.DataFrame(scaled_Mezclados)
X_AlemanNor = pd.DataFrame(scaled_Aleman)
X_PeliculaNor = pd.DataFrame(scaled_Pelicula)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
for i in range(1, 101, 10):
    model = GradientBoostingClassifier(random_state=1, n_estimators=i)
    scores = cross_val_score(model, X_PeliculaNor, y_Pelicula, cv=10, scoring='accuracy')
    print('Numero estimadores' , i)
    print(scores.mean())
```

```python
for i in range(1, 101, 10):
    model = RandomForestClassifier(n_estimators=i)
    scores = cross_val_score(model, X_MezcladosNor, y_Mezclados, cv=10, scoring='accuracy')
    print('Numero estimadores' , i)
    print(scores.mean())
```

```python
for i in [1, 10, 100, 1000]:
    for j in [0.1, 0.01, 0.001, 1]:
        model = SVC(kernel='rbf', C=i, gamma=j)
        scores = cross_val_score(model, X_MezcladosNor, y_Mezclados, cv=10, scoring='accuracy')
        print('C =', i, 'gamma =', j, ':', scores.mean())
```
